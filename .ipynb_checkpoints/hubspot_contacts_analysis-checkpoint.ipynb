```python
import pandas as pd
import numpy as np
import sklearn
import time
```

# Data Loading and Initial Inspection

Loading.

```python
# load the data
data = pd.read_csv('hubspot_data.csv')

# Copy the data to df
df = data.copy()
```

Initial inspection.

```python
# Explore missing values
df.isna().sum().sort_values(ascending=False)
```

# Handling Missing Values

Determening what variables have more than 50% missing data.

```python
# Drop variables with more than 50% missing data
sparse_data = []
for key, value in df.isna().sum().sort_values(ascending=False).to_dict().items():
    if value > (0.5 * len(df)):
        sparse_data.append(key)
print(sparse_data)
```

Removing uneccessary varaibles / variables with missing values that cannot be enriched

```python
# Removing unecessary variables
key_variables = ['Record ID - Contact', 
                 'Bundle Note',
                 'Bundle Lost',
                 'City',
                 'Industry',
                 'Date of last meeting booked in meetings tool',
                 'Sales Stage',
                 'Original Source',
                 'Original Source Drill-Down 1',
                 'Create Date']

# Storing key variables in new DataFrame data
df = df[key_variables]
```

# Initial Visualizations: Leads by Location

Using OpenCage Geocoder to get coordinates from a list of names.

```python
# Get API Key
with open('./opencage_key.txt', 'r') as file:
    key = str(file.read().strip())
```

```python
from opencage.geocoder import OpenCageGeocode
import time

# Initialize geocoder
geocoder = OpenCageGeocode(key)

# Initialize list for coordinates and cities
coords = []
cities = list(df['City'].dropna())  # Make sure 'City' column exists in the df

# Get the coordinates for each city
for city in cities:
    try:
        result = geocoder.geocode(city)
        if result:
            lat = result[0]['geometry']['lat']
            lng = result[0]['geometry']['lng']
            print(f"Success. Appending to dictionary: {lat}, {lng}")
            coords.append({'City': city, 'Latitude': lat, 'Longitude': lng})  # Append to 'coords'
        else:
            coords.append({'City': city, 'Latitude': None, 'Longitude': None})  # Append to 'coords'
            print(f"Not Found. Appending to dictionary: None, None")
    except Exception as e:
        print(f"Error fetching coordinates for {city}: {e}")
    time.sleep(1)  # Respect API rate limits
```

```python
# Store the coords (if first run)
%store coords
```

```python
# Retrieve the coords (if not first run)
%store -r coords
```

Plotting leads by location.

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Store coords as a DF
coords_df = pd.DataFrame(coords)

# Create a new figure and axes with a map projection
fig = plt.figure(figsize=(12, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-130, -60, 20, 55], crs=ccrs.PlateCarree())

# Add features
ax.coastlines(resolution='50m')
ax.add_feature(cfeature.BORDERS, linestyle='-')
ax.add_feature(cfeature.STATES, linestyle=':')
ax.add_feature(cfeature.LAND, color='white')
ax.add_feature(cfeature.OCEAN, color='lightblue')
ax.add_feature(cfeature.LAKES, color='lightblue')

# Plot coordinates and points
lats = coords_df['Latitude']
lons = coords_df['Longitude']
ax.scatter(lons, lats, color='green', marker='o', s=20, alpha=.4, transform=ccrs.PlateCarree(), label='Cities')

# Add title and show the plot
plt.title('City Locations on Map')
plt.legend()
plt.show()
```

# Distribution of Leads

## Distribution of Leads by Platform

The table below shows the count of leads by platform/source.

```python
# Distribution of Leads by Platform
pd.DataFrame(df['Original Source Drill-Down 1'].value_counts())
```

The table below shows the count of sales stages, grouped by platform/source.

```python
# Convert Sales Stage to Dummy Cols
df_dummies = pd.get_dummies(data=df, columns=['Sales Stage'])

# List all the dummy variable columns (those starting with "Sales Stage_")
dummy_columns = [col for col in df_dummies.columns if col.startswith('Sales Stage_')]

# Group by 'Original Source Drill-Down 1' and sum the True values for all dummy columns
grouped = df_dummies.groupby('Original Source Drill-Down 1')[dummy_columns].sum()

# Reorder the columns of grouped
grouped = grouped[['Sales Stage_Closed Won', 
                   'Sales Stage_Onboard', 
                   'Sales Stage_Script Review No Showed',
                   'Sales Stage_Script Review Showed',
                   'Sales Stage_Script Review Scheduled',
                   'Sales Stage_Brief Call No Showed',
                   'Sales Stage_Brief Call Showed',
                   'Sales Stage_Brief Call Scheduled',
                   'Sales Stage_Sign Up'
                  ]]

display(grouped.sort_values(by='Sales Stage_Closed Won', ascending=False))
```
